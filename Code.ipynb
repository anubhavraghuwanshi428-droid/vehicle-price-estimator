#Importing the Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

#Data Collection and Processing

```python
# loading the data from csv file to pandas dataframe
vehicle_data = pd.read_csv("C:\\Users\\siddh\\OneDrive\\Desktop\\ai ml poject\\dataset.csv")
```

```python
# inspecting the first 5 rows of the dataframe
vehicle_data.head()
```

```python
# checking the number of rows and columns
vehicle_data.shape
```

```python
# getting some information about the dataset
vehicle_data.info()
```

```python
# checking the number of missing values
vehicle_data.isnull().sum()
```

```python
# checking the distribution of categorical data
print(vehicle_data.fuel.value_counts())
print(vehicle_data.seller_type.value_counts())
print(vehicle_data.transmission.value_counts())
```

#Encoding the Categorical Data

```python
# encoding the categorical data
pd.set_option("future.no_silent_downcasting", True)
vehicle_data.replace({"fuel": {"Petrol": 0, "Diesel": 1, "CNG": 2,"LPG": 3}}, inplace=True)
vehicle_data.replace({"seller_type": {"Dealer": 0, "Individual": 1,"Trustmark Dealer": 2}}, inplace=True)
vehicle_data.replace({"transmission": {"Manual": 0, "Automatic": 1}}, inplace=True)
```

```python
vehicle_data.head()
```

#Splitting the data and Target

```python
X = vehicle_data.drop(['name','selling_price'],axis=1)
Y = vehicle_data['selling_price']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

Model Training

#1. Linear Regression

```python
# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

#Model Evaluation

```python
# prediction on Training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

#Visualize the actual prices and Predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
